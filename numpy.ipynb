```python
import numpy as np

np.random.seed(85)

result = []

for i in range(3):
  a = np.arange(1,46)
  np.random.shuffle(a)
  lotto = a[:6]
  result.append(lotto)

result_array = np.array(result)

print('第三轮乐透号码结果（列表）:\n', result)
print('\n二维数组 (shape (3, 6)):\n', result_array)
print('数组形状：', result_array.shape)
```
